```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0,10,11)
b = a ** 4 
```

```python
x = np.arange(0,10)
y = 2 * x
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 1) 
# axes[0][0].plot(x,y)
# axes[1][1].plot(a,b)

for ax in axes:
    ax.plot(x,y)
    
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (4,10), dpi = 200) 
axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].set_ylabel('Y label')
axes[1][1].plot(a,b)
axes[1][1].set_title('TITLE 1.1')

fig.suptitle('Figure level title')
plt.tight_layout()

fig.set_figwidth(10)
fig.savefig('new_subplots.png', bbox_inches = 'tight')
```
