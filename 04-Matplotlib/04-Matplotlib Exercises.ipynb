**В этом упражнении у Вас будут данные для кривой доходности для двух различных моментов времени. И Вы построите графики на основе этих данных.**

## Задание 2, часть 1: Данные для кривой доходности

**Мы взяли для Вас нужные данные для кривой доходности с сайта [US Treasury Dept.](https://www.treasury.gov/resource-center/data-chart-center/interest-rates/pages/textview.aspx?data=yield) Данные показывают процентные ставки облигаций США для различных сроков контракта. Список labels отображает сроки контракта для соответствующих позиций в списках значений.**

**Задание: Запустите ячейку ниже, чтобы создать списки для последующего рисования графика.**

```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

july16_2007 =[4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

**Задание: выясните, как можно нарисовать обе эти линии на одном и том же графике. Добавьте на график легенду.**

```python
# ПИШИТЕ ВАШ КОД ЗДЕСЬ
fig = plt.figure(figsize = (5,3))

ax = fig.add_axes([0,0,1,1])


ax.plot(labels,july16_2007, label = 'july16_2007')

ax.plot(labels,july16_2020, label = 'july16_2020')
ax.legend(loc = (0.7, 0.5))
```

```python
# НЕ ЗАПУСКАЙТЕ ЯЧЕЙКУ НИЖЕ, ИНАЧЕ ГРАФИК ИСЧЕЗНЕТ!
```

**Задание: на графике выше легенда выглядит несколько странно - хотя она и не загораживает график, было бы лучше поместить легенду *снаружи* графика. Найдите способ переместить легенду так, чтобы она оказалась снаружи.**

```python
# ПИШИТЕ ВАШ КОД ЗДЕСЬ
fig = plt.figure(figsize = (5,3))

ax = fig.add_axes([0,0,1,1])


ax.plot(labels,july16_2007, label = 'july16_2007')

ax.plot(labels,july16_2020, label = 'july16_2020')
ax.legend(loc = (1.1, 0.5))
```

```python
# НЕ ЗАПУСКАЙТЕ ЯЧЕЙКУ НИЖЕ, ИНАЧЕ ГРАФИК ИСЧЕЗНЕТ!
```

**Задание: хотя на графике ясно видно разницу между 2007 и 2020 годами, когда мы наносим эти линии на один и тот же график, нам становится сложнее увидеть детали внутри каждого отдельного года. С помощью .suplots() нарисуйте графики ниже, в которых эти детали будут лучше видны.**

```python
# ПИШИТЕ ВАШ КОД ЗДЕСЬ
fig, axes = plt.subplots(nrows=2, ncols=1)
axes[0].plot(labels,july16_2007)
axes[0].set_title('july16_2007')

axes[1].plot(labels,july16_2020)
axes[1].set_title('july16_2020')

plt.tight_layout()
```

```python
# НЕ ЗАПУСКАЙТЕ ЯЧЕЙКУ НИЖЕ, ИНАЧЕ ГРАФИК ИСЧЕЗНЕТ!
```

**Дополнительное бонусное задание: попробуйте воспроизвести график ниже, с двумя осями. Хотя в данном примере две разные оси по сути не нужны, это хорошее упражнение по Matplotlib, которое пригодится в других ситуациях.**

```python
# ПИШИТЕ ВАШ КОД ЗДЕСЬ
fig, axes = plt.subplots()
axes.spines['left'].set_color('blue')
axes.spines['left'].set_linewidth(2)
axes.plot(labels,july16_2007, color = 'blue')
axes.set_ylabel('2007', fontsize=11, color="blue")
for label in axes.get_yticklabels():
    label.set_color("blue")


ax2 = axes.twinx()
ax2.spines['right'].set_color('red')
ax2.spines['right'].set_linewidth(2)
ax2.plot(labels,july16_2020, color = 'red')
ax2.set_ylabel('2020', fontsize=11, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")

```
